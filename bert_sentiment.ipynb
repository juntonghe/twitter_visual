```python
import tensorflow as tf
mnist = tf.keras.datasets.mnist

(x_train, y_train),(x_test, y_test) = mnist.load_data()
x_train, x_test = x_train / 255.0, x_test / 255.0

model = tf.keras.models.Sequential([
  tf.keras.layers.Flatten(input_shape=(28, 28)),
  tf.keras.layers.Dense(128, activation='relu'),
  tf.keras.layers.Dropout(0.2),
  tf.keras.layers.Dense(10, activation='softmax')
])

model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

model.fit(x_train, y_train, epochs=1)
model.evaluate(x_test, y_test)
```

# import training set and test set

```python
import pandas as pd
df_dataset = pd.read_csv('/content/tweet_dataset.csv')
df_dataset.head()
```

```python
df_dataset_cleared = df_dataset.dropna().sample(frac=1).reset_index(drop=True)
print( len(df_dataset_cleared), len(df_dataset))
df_dataset_cleared.head()
```

```python
df_dataset_cleared['new_sentiment'].value_counts()
```

## label encode to integer

```python
from sklearn import preprocessing
le = preprocessing.LabelEncoder()
```

```python
label_num_array = le.fit_transform(df_dataset_cleared['new_sentiment'])
label_num_array[:10]
```

```python
df_dataset_cleared['label_num'] = label_num_array
df_dataset_cleared.head()
```

## changing the dataset to the standard format of simpletransformer input

---



```python
df_dataset_cleared[['text','label_num']].head()
```

```python
train_df = df_dataset_cleared[['text','label_num']][:22000]
train_df.columns = ["text", "labels"]
eval_df = df_dataset_cleared[['text','label_num']][22000:]
eval_df.columns = ["text", "labels"]
```

```python
train_df.head()
```

```python
from simpletransformers.classification import ClassificationModel, ClassificationArgs
import pandas as pd
import logging


logging.basicConfig(level=logging.INFO)
transformers_logger = logging.getLogger("transformers")
transformers_logger.setLevel(logging.WARNING)
```

# create bert three-classification model

```python
# Optional model configuration
model_args = ClassificationArgs(num_train_epochs=1,overwrite_output_dir=True)

# Create a ClassificationModel
model = ClassificationModel(
    'bert',
    'bert-base-cased',
    num_labels=3,
    args=model_args
)
```

# training the model

```python
# Train the model
model.train_model(train_df)
```

# evaluation

```python
# Evaluate the model
result, model_outputs, wrong_predictions = model.eval_model(eval_df)

result, model_outputs, wrong_predictions 
```

## get the result: predict_label

```python
import numpy as np
predict_label = np.argmax(model_outputs,axis=1)
model_outputs,predict_label
```

## get the classification report


```python
from sklearn.metrics import classification_report
```

```python
label_true = eval_df['labels']
label_true[:5]
```

```python
print(classification_report(label_true, predict_label))
```

## save the result to .csv file

```python
eval_df['predict_label'] = predict_label
eval_df
```

```python
eval_df.to_csv('predict_result.csv')
```
